```python
import geopandas as gpd
```

```python
states_file = gpd.read_file("../data/state/")
```

```python
states_file.plot()
```

```python
states_file.crs = {"init": "epsg:4326" }
```

```python
states_file.plot()
```

```python
states = states_file[ 
    ~states_file["STATEFP"].str.contains(
        "72|66|60|02|78|15|69"
    )
].to_crs(
    {'init':'epsg:2163'}
)

states.plot()
```

```python
states.head(1)
```

```python
states[
    ['STATEFP',
     'STATENS', 
     'GEOID', 
     'STUSPS',
     'NAME', 
     'LSAD', 
     'ALAND',
     'AWATER',
     'INTPTLAT',
     'INTPTLON',
     'geometry']
].to_file("../output/states_simple/states_simple.shp")
# Note that the deprecation warning will go away with next geopandas release: https://github.com/geopandas/geopandas/issues/882
```

```python
ax = gpd.read_file("../output/states_simple/states_simple.shp").plot(
    figsize = (16, 9),
    edgecolor = "white",
    linewidth = 0.5,
    color = "steelblue",
    alpha = 0.8
)

ax.set_title("What a lovely map", );
```
